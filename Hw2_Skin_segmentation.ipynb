## Чтение выборки

Давайте посмотрим на выборку и попробуем понять, какой смысл имеют

```python
dataset_path = "Skin_NonSkin.txt"
dataset_file = open(dataset_path, 'r')
dataset_text = dataset_file.read().split('\n')
```

```python
dataset = [list(map(int, line.split('\t'))) for line in dataset_text if len(line) > 0]
#print(dataset[:10])
#print(len(dataset))
```

```python
X = [line[-2::-1] for line in dataset] #the only line modified
y = [line[-1] for line in dataset] 
```

```python
import numpy as np

np.random.seed(0)
indices = np.random.permutation(len(X))
X = np.array(X)
y = np.array(y)
test_size = 100000
X_train = X[indices[:-test_size]]
y_train = y[indices[:-test_size]]
X_test  = X[indices[-test_size:]]
y_test  = y[indices[-test_size:]]
```

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier()
model.fit(X_train, y_train)
```

```python
predicted = model.predict(X_test)
print(predicted[:10])
```

```python
print(sum(y_test == predicted)/test_size)
```

```python
def print_ljust(l):
    for el in l:
        print(str(el).ljust(24), end='')
    print()
    return

def get_quality(model):
    model.fit(X_train, y_train)
    predicted = model.predict(X_test)
    return sum(y_test == predicted)/test_size

print('#  ', end='')
print_ljust(['[euclidean, uniform]', '[euclidean, distance]', '[manhattan, uniform]', '[manhattan, distance]'])
for i in range(1, 11):
    values = []
    values.append(get_quality(KNeighborsClassifier(n_neighbors=i)))
    values.append(get_quality(KNeighborsClassifier(n_neighbors=i, weights='distance')))
    values.append(get_quality(KNeighborsClassifier(n_neighbors=i, p=1)))
    values.append(get_quality(KNeighborsClassifier(n_neighbors=i, p=1, weights='distance')))
    
    print(str(i).ljust(5), end='')
    print_ljust(values)
    
```

И наш победитель - KNN с весами относительно евклидового расстояния.
